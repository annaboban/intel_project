```python
from IPython import get_ipython
from IPython.display import display
# %%
# Install necessary libraries
!pip install qrcode pandas openpyxl Pillow pytesseract

# Install Tesseract OCR engine
!apt-get install tesseract-ocr

import pandas as pd
import qrcode
from IPython.display import display, Image
from google.colab import files
import os # Import the os module to get the file extension
from PIL import Image as PILImage, ImageDraw, ImageFont # Import Pillow modules
import pytesseract # Import pytesseract
import tempfile # Import tempfile for temporary files
import sqlite3 # Import sqlite3 for database operations

# Set the path to the tesseract executable (this might vary depending on your installation)
# If running in Colab, this path is usually correct.
pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# --- Database Setup ---
db_name = 'label.db'
conn = None # Initialize connection to None

try:
    # Create or connect to the SQLite database
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    # Create a table to store original and OCR text
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS labels (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        original_text TEXT,
        ocr_text TEXT
    )
    ''')

    # Commit the creation of the table
    conn.commit()
    print(f"Database '{db_name}' and table 'labels' set up successfully.")

except sqlite3.Error as e:
    print(f"Database error: {e}")
    # Handle the error appropriately, maybe exit or skip database operations
    conn = None # Ensure conn is None if setup fails

# Upload your dataset file
uploaded = files.upload()

# Assuming the uploaded file is one of the keys in the uploaded dictionary
file_name = list(uploaded.keys())[0]

df = None # Initialize df to None

# Attempt to read the file using different pandas readers
try:
    print(f"Attempting to read {file_name} as CSV...")
    df = pd.read_csv(file_name)
    print("Successfully read as CSV.")
except Exception as e_csv:
    print(f"Could not read as CSV: {e_csv}")
    try:
        print(f"Attempting to read {file_name} as Excel...")
        # Determine the engine based on file extension if possible
        file_extension = os.path.splitext(file_name)[1].lower()
        if file_extension == '.xlsx':
            df = pd.read_excel(file_name, engine='openpyxl')
        else:
             df = pd.read_excel(file_name) # pandas might guess the engine
        print("Successfully read as Excel.")
    except Exception as e_excel:
        print(f"Could not read as Excel: {e_excel}")
        try:
            print(f"Attempting to read {file_name} as JSON...")
            df = pd.read_json(file_name)
            print("Successfully read as JSON.")
        except Exception as e_json:
            print(f"Could not read as JSON: {e_json}")
            print(f"Error: Could not read {file_name} as a recognized data file type (CSV, Excel, or JSON).")
            print("Please ensure the uploaded file is a valid data file in one of these formats.")

# Proceed only if the dataframe was successfully created
if df is not None:
    # Generate labels and QR codes for each row
    for index, row in df.iterrows():
        # Create a string with the details for the label
        label_details = ""
        for col in df.columns:
            # Ensure values are strings to avoid TypeError during concatenation
            label_details += f"{col}: {str(row[col])}\n"

        # Generate a QR code from the label details
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_L,
            box_size=10,
            border=4,
        )
        # Ensure the data added to QR code is a string
        qr.add_data(str(label_details))
        qr.make(fit=True)

        # Create an image from the QR code
        qr_img = qr.make_image(fill_color="black", back_color="white").convert('RGB') # Convert to RGB for pasting

        # --- Combine Label Details and QR Code into a single image ---

        # Determine dimensions for the combined image
        qr_width, qr_height = qr_img.size

        # Estimate text height (this is a rough estimate, could be improved)
        # You might need to adjust font size and line height based on your needs
        try:
            font = ImageFont.truetype("arial.ttf", 15) # Try to use a common font
        except IOError:
            font = ImageFont.load_default() # Fallback to default font

        # Calculate text size using textbbox
        dummy_img = PILImage.new('RGB', (1, 1))
        dummy_draw = ImageDraw.Draw(dummy_img)
        bbox = dummy_draw.textbbox((0, 0), label_details, font=font)
        text_width = bbox[2] - bbox[0]
        text_height = bbox[3] - bbox[1]


        # Add some padding
        padding = 20
        total_width = max(qr_width, text_width + padding * 2)
        total_height = qr_height + text_height + padding * 3 # Add space for text above QR

        # Create a new blank image
        combined_img = PILImage.new('RGB', (total_width, total_height), color='white')
        draw = ImageDraw.Draw(combined_img)

        # Paste the QR code onto the combined image
        qr_paste_x = (total_width - qr_width) // 2 # Center the QR code
        qr_paste_y = text_height + padding * 2
        combined_img.paste(qr_img, (qr_paste_x, qr_paste_y))

        # Draw the text onto the combined image
        text_paste_x = padding
        text_paste_y = padding
        draw.multiline_text((text_paste_x, text_paste_y), label_details, fill="black", font=font)

        # Display the combined label image
        print(f"Displaying Label for Row {index}:")
        display(combined_img)
        print("-" * 20) # Separator for each label

        # --- Perform OCR on the combined image ---
        ocr_text = "" # Initialize ocr_text
        print("Performing OCR on the generated image...")
        try:
            # Save the image to a temporary file
            with tempfile.NamedTemporaryFile(suffix=".png", delete=False) as tmpfile:
                combined_img.save(tmpfile.name)
                temp_filepath = tmpfile.name

            # Use pytesseract to extract text from the image
            ocr_text = pytesseract.image_to_string(PILImage.open(temp_filepath))

            # Print the original details and the OCR generated text
            print("\nOriginal Details:")
            print(label_details)
            print("\nOCR Generated Text:")
            print(ocr_text)

            # Clean up the temporary file
            os.remove(temp_filepath)

        except pytesseract.TesseractNotFoundError:
            print("\nError: Tesseract is not installed or not in your PATH.")
            print("Please install Tesseract OCR engine to perform OCR.")
        except Exception as e:
            print(f"\nAn error occurred during OCR processing: {e}")

        # --- Insert data into the database ---
        if conn is not None: # Only attempt to insert if database connection is valid
            try:
                cursor = conn.cursor()
                cursor.execute("INSERT INTO labels (original_text, ocr_text) VALUES (?, ?)", (label_details, ocr_text))
                conn.commit()
                print("Data inserted into database.")
            except sqlite3.Error as e:
                print(f"Error inserting data into database: {e}")

        print("=" * 40) # Separator for each row's output

# --- Close database connection after processing all rows ---
if conn is not None:
    conn.close()
    print("Database connection closed.")
from google.colab import files

# 'label.db' is the name of your database file
database_file_name = 'label.db'

try:
  files.download(database_file_name)
  print(f"'{database_file_name}' downloaded successfully.")
except FileNotFoundError:
  print(f"Error: '{database_file_name}' not found. Please ensure the file was created.")
except Exception as e:
  print(f"An error occurred during download: {e}")
```

```python
#Install necessary libraries
!pip install Pillow pytesseract pandas

# Install Tesseract OCR engine
!apt-get install tesseract-ocr

import pytesseract
from PIL import Image as PILImage
from google.colab import files
import os
import sqlite3
import datetime # Import the datetime module
import pandas as pd # Import pandas for displaying the database
from IPython.display import display # Import display for pretty printing in Colab

# Set the path to the tesseract executable (this might vary depending on your installation)
pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# --- Database Setup ---
# CHANGE DATABASE NAME HERE
db_name = 'scanned_labels1.db'
conn = None # Initialize connection to None

try:
    # Create or connect to the SQLite database
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    cursor.execute('''
    CREATE TABLE IF NOT EXISTS ocr_results (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        filename TEXT,
        ocr_text TEXT,
        scan_timestamp DATETIME
    )
    ''')

    # Commit the creation of the table
    conn.commit()
    # Print the new database name in the success message
    print(f"Database '{db_name}' and table 'ocr_results' set up successfully.")

except sqlite3.Error as e:
    print(f"Database error during setup: {e}")
    # Handle the error appropriately, maybe exit or skip database operations
    conn = None # Ensure conn is None if setup fails

print("You can upload label image files one after another. Press 'Cancel' in the upload dialog when you are finished.")

processed_files_count = 0

while True:
    print(f"\nUploading file {processed_files_count + 1}...")
    uploaded = files.upload()

    if not uploaded:
        print("\nNo file uploaded. Assuming you are finished.")
        break # Exit the loop if no file is uploaded (e.g., user cancelled)

    # Get the name of the first (and likely only) file uploaded in this iteration
    file_name = list(uploaded.keys())[0]
    print(f"Processing file: {file_name}")

    ocr_text = "" # Initialize ocr_text for this iteration
    scan_timestamp = datetime.datetime.now() # Get the current date and time

    try:
        # Open the image file using Pillow
        img = PILImage.open(file_name)

        # Perform OCR on the image
        ocr_text = pytesseract.image_to_string(img)

        # Print the extracted text
        print("\nOCR Extracted Text:")
        print(ocr_text)
        print("-" * 30) # Separator

        # Increment the count only for files that were processed
        processed_files_count += 1


    except pytesseract.TesseractNotFoundError:
        print("\nError: Tesseract is not installed or not in your PATH.")
        print("Please install Tesseract OCR engine to perform OCR.")
        # Break the loop if Tesseract is not found, as OCR won't work
        break
    except Exception as e:
        print(f"\nAn error occurred while processing {file_name}: {e}")
        # Continue to the next upload if an error occurs during processing
        continue # Skip database insertion for this file if OCR failed

    finally:
        # Insert data into the database after attempting OCR
        # This finally block will execute even if an error occurred in try/except
        # BUT if 'continue' was used in except, this block is skipped.
        # So, we need to ensure 'ocr_text' was successfully generated or handle its absence.
        # The check for conn is vital.
        if conn is not None and ocr_text: # Only insert if conn is valid and ocr_text is not empty
            try:
                cursor = conn.cursor()
                # Use the filename, extracted text, and timestamp
                cursor.execute("INSERT INTO ocr_results (filename, ocr_text, scan_timestamp) VALUES (?, ?, ?)",
                               (file_name, ocr_text, scan_timestamp))
                conn.commit()
                print("Data inserted into database.")
            except sqlite3.Error as e:
                print(f"Error inserting data into database: {e}")

        # Optional: Clean up the uploaded file from the Colab environment after processing
        # os.remove(file_name) # Uncomment this line if you want to delete the file
        pass # Use pass if no cleanup is needed in finally


# --- Close database connection after processing all rows ---
if conn is not None:
    conn.close()
    print("\nDatabase connection closed after processing.")

print(f"\nFinished processing a total of {processed_files_count} uploaded images.")

# --- Code to Display the Database ---
print("\n--- Displaying Database Contents ---")

# Need to re-establish connection to read the database
conn_display = None
try:
    # Connect to the new database name
    conn_display = sqlite3.connect(db_name)
    cursor_display = conn_display.cursor()

    # Select all data from the ocr_results table
    cursor_display.execute("SELECT * FROM ocr_results")

    # Fetch all the results
    rows_display = cursor_display.fetchall()

    # Get the column names from the cursor description
    column_names_display = [description[0] for description in cursor_display.description]

    # Check if there are any rows
    if rows_display:
        # Print the new database name in the display message
        print(f"Contents of the '{db_name}' database:")
        # Convert the results to a pandas DataFrame for better display
        df_results = pd.DataFrame(rows_display, columns=column_names_display)

        # Display the DataFrame in Colab
        display(df_results)
    else:
        print(f"No data found in the 'ocr_results' table of '{db_name}'.")

except sqlite3.Error as e:
    print(f"Database error during display: {e}")
except FileNotFoundError:
    # Ensure the error message mentions the new database name
    print(f"Error: Database file '{db_name}' not found. Please ensure the file was created and exists.")
except Exception as e:
    print(f"An unexpected error occurred during display: {e}")

finally:
    # Close the display database connection if it was successfully opened
    if conn_display:
        conn_display.close()
        print("\nDisplay database connection closed.")

# Optional: Download the database file - moved here to be after processing and display
# Ensure the download uses the new database name
print(f"\nDownloading the database file '{db_name}'...")
try:
  files.download(db_name)
  print(f"'{db_name}' downloaded successfully.")
except FileNotFoundError:
  # Ensure the error message mentions the new database name
  print(f"Error: '{db_name}' not found after processing. Ensure data was inserted.")
except Exception as e:
  print(f"An error occurred during download: {e}")
```

```python
# Install necessary libraries
!pip install Pillow pyzbar pandas

# Install the zbar shared library
!apt-get update && apt-get install -y libzbar0

import pyzbar.pyzbar as pyzbar
from PIL import Image as PILImage
from google.colab import files
import io # Import io for handling binary data
import sqlite3 # Import sqlite3 for database operations
import datetime # Import the datetime module
import pandas as pd # Import pandas for displaying the database
from IPython.display import display # Import display for pretty printing in Colab

# --- Database Setup ---
db_name = 'qr_scan_results.db'
conn = None # Initialize connection to None

try:
    # Create or connect to the SQLite database
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    # Create a table to store QR code results
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS qr_scans (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        filename TEXT,
        qr_data TEXT,
        scan_timestamp DATETIME
    )
    ''')

    # Commit the creation of the table
    conn.commit()
    print(f"Database '{db_name}' and table 'qr_scans' set up successfully.")

except sqlite3.Error as e:
    print(f"Database error during setup: {e}")
    # Handle the error appropriately, maybe exit or skip database operations
    conn = None # Ensure conn is None if setup fails

print("You can upload multiple image files containing QR codes. Press 'Cancel' in the upload dialog when you are finished.")

processed_files_count = 0

# Start a loop that continues until the user cancels the upload
while True:
    print(f"\nUploading image file {processed_files_count + 1}...")

    # Upload the image file
    # files.upload() will return an empty dictionary if the user cancels the upload dialog
    uploaded = files.upload()

    # Check if a file was uploaded in this iteration
    if not uploaded:
        print("\nNo file uploaded. Assuming you are finished.")
        break # Exit the loop if no file is uploaded (user likely clicked Cancel)

    # Process each uploaded file in this batch
    for file_name, file_content in uploaded.items():
        print(f"Processing file: {file_name}")

        scan_timestamp = datetime.datetime.now() # Get the current date and time for this scan
        qr_data_list = [] # Use a list to store data from potentially multiple QR codes per image

        try:
            # Open the image file using Pillow from its content (bytes)
            img = PILImage.open(io.BytesIO(file_content))

            # Find QR codes in the image
            decoded_objects = pyzbar.decode(img)

            # Check if any QR codes were found
            if decoded_objects:
                print(f"\nFound {len(decoded_objects)} QR code(s) in '{file_name}':")
                # Iterate over the found QR codes
                for obj in decoded_objects:
                    # Decode the QR code data (it's in bytes, so decode to string)
                    qr_data = obj.data.decode('utf-8')
                    print("Decoded QR Code Details:")
                    print(qr_data)
                    print("-" * 20) # Separator for multiple QR codes or different files
                    qr_data_list.append(qr_data) # Add decoded data to the list

            else:
                print(f"\nNo QR codes found in the image '{file_name}'.")
                qr_data_list.append("") # Add an empty string if no QR code is found

            # --- Insert data into the database ---
            if conn is not None: # Only attempt to insert if database connection is valid
                try:
                    cursor = conn.cursor()
                    # If there were multiple QR codes, insert each one as a separate row.
                    # If no QR codes or only one, the loop below handles it.
                    if qr_data_list:
                        for data_item in qr_data_list:
                            cursor.execute("INSERT INTO qr_scans (filename, qr_data, scan_timestamp) VALUES (?, ?, ?)",
                                           (file_name, data_item, scan_timestamp))
                        conn.commit()
                        print(f"{len(qr_data_list)} QR data entries inserted into database for {file_name}.")
                    else:
                         # Should not happen if qr_data_list is initialized with "" when no QR code found, but added for safety
                         print(f"No data to insert for {file_name}.")

                except sqlite3.Error as e:
                    print(f"Error inserting QR data into database for {file_name}: {e}")


            # Increment the count for each file successfully processed
            processed_files_count += 1

        except Exception as e:
            print(f"\nAn error occurred while processing '{file_name}': {e}")
            # Continue to the next file if an error occurs with the current one
            continue


# Code execution continues here after the loop finishes
print(f"\nFinished processing a total of {processed_files_count} uploaded images.")

# --- Close database connection after processing all rows ---
if conn is not None:
    conn.close()
    print("\nDatabase connection closed after processing.")

# --- Code to Display the Database ---
print("\n--- Displaying Database Contents ---")

# Need to re-establish connection to read the database
conn_display = None
try:
    # Connect to the database
    conn_display = sqlite3.connect(db_name)

    # Read the data from the 'qr_scans' table into a pandas DataFrame
    df_results = pd.read_sql_query("SELECT * FROM qr_scans", conn_display)

    # Check if the DataFrame is empty
    if not df_results.empty:
        print(f"Contents of the '{db_name}' database:")
        # Display the DataFrame in Colab
        display(df_results)
    else:
        print(f"No data found in the 'qr_scans' table of '{db_name}'.")

except sqlite3.Error as e:
    print(f"Database error during display: {e}")
except FileNotFoundError:
    print(f"Error: Database file '{db_name}' not found. Please ensure the file was created.")
except Exception as e:
    print(f"An unexpected error occurred during display: {e}")

finally:
    # Close the display database connection if it was successfully opened
    if conn_display:
        conn_display.close()
        print("\nDisplay database connection closed.")

# Optional: You might want to download the database file after displaying
# from google.colab import files
# try:
#     files.download(db_name)
#     print(f"'{db_name}' downloaded successfully.")
# except FileNotFoundError:
#     print(f"Error: '{db_name}' not found after processing. Ensure data was inserted.")
# except Exception as e:
#     print(f"An error occurred during download: {e}")
```
